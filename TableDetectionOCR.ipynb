```python
from pprint import pprint
from paddlenlp import Taskflow
import pandas as pd
import re

df = pd.read_csv("text.csv")
data = df.iloc[:,2]
single_quote_pattern = r"'(.*?)'"
double_quote_pattern = r'"(.*?)"'
extracted_content = data.apply(lambda x: (re.search(single_quote_pattern, x) and re.search(single_quote_pattern, x).group(1)) or (re.search(double_quote_pattern, x) and re.search(double_quote_pattern, x).group(1)) or '')
print(str(extracted_content))
merged_content = ' '.join(extracted_content)
print(str(merged_content))

schema = ['person', 'date', 'phone number', 'e-mail', 'company', 'discription', 'destination', 'weights', 'fax number', \
        'goods']
ie_en = Taskflow('information_extraction', schema=schema, model='uie-m-base')
pprint(ie_en(str(merged_content)))
```
